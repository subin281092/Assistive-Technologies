```python
import cv2
import torch
import pyttsx3
import speech_recognition as sr

training = torch.hub.load("ultralytics/yolov5", "yolov5s")
speak = pyttsx3.init()
recognizer = sr.Recognizer()

def get_voice_input():
    
    with sr.Microphone() as source:
        print("Say the object class you want to detect:")
        speaktext="Say the object class you want to detect:"
        speak.say(speaktext)
        speak.runAndWait()
        audio = recognizer.listen(source)
        text = recognizer.recognize_google(audio)
        print(f"Recognized: {text}")
        return text.strip().lower()
       
# Get the object class from voice input
target_class = get_voice_input()

if target_class is None:
    print("Failed to get a valid object class. Exiting.")
    exit()

# Initialize video capture
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print("Error: Could not open video camera")
else:
    print("Camera opened")

fps = cap.get(cv2.CAP_PROP_FPS)

while True:
    ret, imcap = cap.read()
    
    detect = training(imcap) 
    out = detect.render()[0].copy()
    detectedob = detect.pandas().xyxy[0]

    # Checking if target class is detected
    detected_classes = detectedob['name'].str.lower().unique()
    if target_class in detected_classes:
        speakobject = f"{target_class} detected"
    else:
        speakobject = f"No {target_class} detected"

    # Display the frame with detection information
    cv2.putText(out, f'FPS: {fps:.2f}', (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)
    cv2.putText(out, speakobject, (10, 70), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    cv2.imshow("test", out)

    # Speak the detection
    speak.say(speakobject)
    speak.runAndWait()

    
    if cv2.waitKey(1) == ord('x'):
        break

cap.release()
cv2.destroyAllWindows()
```
